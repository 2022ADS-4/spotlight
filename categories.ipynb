```python
import pandas as pd
```

```python
data = pd.read_csv('data/ml-data.csv')
data
```

```python
data['title'].nunique()
```

```python
data['genres'].nunique()
```

```python
data['genres'] = data['genres'].fillna('no genres listed')
```

```python
genres_for_movie = []
genres_for_movie = data['genres'].str.split('|')
genres_for_movie
```

```python
genres_columns = []
for element in genres_for_movie:
    if len(element) > 0:
        for el in element:
            #print(el)
            genres_columns.append(el)

genres_columns = set(genres_columns)
genres_columns
```

```python
for el in genres_columns:
    data[el] = 0
data
```

```python
for element in genres_columns:
    for i in range(len(genres_for_movie)):
        if element in genres_for_movie[i]:
            data.at[i,element] = 1
data
```

```python
import numpy as np
data['no genres listed'] = np.where(data['(no genres listed)'] == 1, data['(no genres listed)'], data['no genres listed'])
```

```python
genres_columns.remove('(no genres listed)')
data.drop(['(no genres listed)'], axis=1)
```

```python
import seaborn as sns
categories = data.melt(value_vars=list(genres_columns))
categories_to_plot = categories[categories["value"] != 0]
sns.countplot(data=categories_to_plot, y="variable")
```
